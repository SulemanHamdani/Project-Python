```python
#load new_data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
data = pd.read_csv('final_data.csv')
```

```python
data.head()
```

# Religion vs No Religion

```python
colors = ['Yes', 'No', 'To some extent', "I’ve never heard about the Aurat March."]
palette_colors = sns.color_palette('Set2', 4)
palette_dict = {continent: color for continent, color in zip(colors, palette_colors)}
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Religion', hue='1.1 Do you think religion provides the complete code of conduct of a person’s life?', data=data, palette=palette_dict)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Religion', hue='1.2 Do you think religious freedom is everyone’s right? ', data=data, palette=palette_dict)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Religion', hue='1.4 Do you think people who wear modest clothes appear to be more moral.', data=data, palette=palette_dict)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Religion', hue='2.8 Do you think concerts/parties are spreading vulgarity in our society?', data=data, palette=palette_dict)
```

## Females who believe patriarchy is not necessary

```python

females = data
females = females[data['Gender'] == 'Female']
females_patriarchy_no = females
females_patriarchy_no = females_patriarchy_no[females['2.5 Do you believe patriarchy is a necessary/effective type of social system?'] == 'No']

```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Gender', hue='2.8 Do you think concerts/parties are spreading vulgarity in our society?', data=females_patriarchy_no, palette=palette_dict)
```
