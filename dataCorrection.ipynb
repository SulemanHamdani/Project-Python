```python
import pandas as pd
```

```python
#load data.csv
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
#remove the column 'Timestamp'
data = data.drop('Timestamp', axis=1)
```

```python
data.head()
```

```python
#In the relgion column, if religion is none or None, or includes the word agnostic or atheist, make it atheist

data['Religion'] = data['Religion'].replace(['None', 'none'], 'Atheist')
data['Religion'] = data['Religion'].replace(['Agnostic', 'agnostic'], 'Atheist')
data['Religion'] = data['Religion'].replace(['Agnostic / Atheist'], 'Atheist')
data['Religion'] = data['Religion'].replace(['Athiest'], 'Atheist')
```

```python
#In the relgion column, if religion is islam or Islam or muslim or Muslim, make it Islam

data['Religion'] = data['Religion'].replace(['Islam', 'islam'], 'Islam')
data['Religion'] = data['Religion'].replace(['Muslim', 'muslim'], 'Islam')
data['Religion'] = data['Religion'].replace(['Islam sunni'], 'Islam')
data['Religion'] = data['Religion'].replace(['Islan'], 'Islam')
data['Religion'] = data['Religion'].replace(['Hindusiam '], 'Hinduism')
data['Religion'] = data['Religion'].replace(['Hinduism '], 'Hinduism')
data['Religion'] = data['Religion'].replace(['Islam '], 'Islam')
```

```python
data['Religion'] = data['Religion'].replace(['Catholic'], 'Christian')
```

```python
data.head()
```

```python
# Make a bar graph of the religion column where it says Atheist vs 1.1 Do you think religion provides the complete code of conduct of a person’s life?

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Religion', hue='1.1 Do you think religion provides the complete code of conduct of a person’s life?', data=data)

```

```python
# Find people who are atheist

atheist = data[data['Religion'] == 'Atheist']
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Religion', hue='1.1 Do you think religion provides the complete code of conduct of a person’s life?', data=atheist)
```
